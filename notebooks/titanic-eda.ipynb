```python
import numpy as np
import pandas as pd
from os import path
import data_helper as dh
```

```python
train_data_file = path.join(path.pardir,'data','external','titanic-train.csv')
test_data_file = path.join(path.pardir,'data','external','titanic-test.csv')
df_train = pd.read_csv(train_data_file,index_col='PassengerId')
df_test = pd.read_csv(test_data_file,index_col='PassengerId')
df_test['Survived']=-888
df = pd.concat((df_train,df_test),axis=0)
```

```python
df.info()
# Columns to be adjusted for null
# 1: Age, 2 : Cabin, Fare, Embarked
```

```python
df['Title'] = df.Name.map(lambda x: dh.get_title(x))
columns = ['Survived','Title','Name','Sex','Age', 'SibSp','Parch','Pclass', 'Fare', 'Ticket', 'Cabin','Embarked']
df = df[columns]
```
